```python
import pandas as pd
```

```python
CLE = pd.read_csv('D:\\O2C QDI Data\\CLE.csv')
```

```python
CLE.rename(columns = {'Entry No_':'Entry_No'}, inplace = True)
```

```python
DCLE= pd.read_csv('D:\\O2C QDI Data\\DCLE.csv')
```

```python
DCLE.rename(columns = {'Cust_ Ledger Entry No_':'Entry_No'}, inplace = True)
```

```python
#Left join
Left_Join_DCLE_CLE = DCLE.merge(CLE, on='Entry_No', how='left')
```

```python
# Replace null values with None
Left_Join_DCLE_CLE['Currency Code'] = Left_Join_DCLE_CLE['Currency Code'].where(pd.notnull(Left_Join_DCLE_CLE['Currency Code']), None)
```

```python
#If condition
Left_Join_DCLE_CLE['Original_Amount'] = Left_Join_DCLE_CLE.apply(lambda x:x['Amount (LCY)'] if x['Entry Type'] == 1 else None, axis=1)
```

```python
#Flag
Left_Join_DCLE_CLE['Flag'] = 'CLE'
```

```python
#Drop column
Left_Join_DCLE_CLE.drop(['Document No__y'], axis=1,inplace = True)
```

```python
Left_Join_DCLE_CLE
```

```python
# Rename Column name
Left_Join_DCLE_CLE.rename(columns = {'Document No__x':'Document_No',
'Sell-To_Customer_No':'Sell_To_Customer_No',
'Posting Date_x':'DCLE_Posting_Date',
'Posting Date_y':'Posting_Date',
'Closed by Amount':'Closed_by_Amount',
'Closed at Date':'Closed_at_Date',
'Closed by Entry No_':'Closed_by_Entry_No',
'Customer Posting Group':'Customer_Posting_Group',
'Salesperson Code':'Salesperson_Code',
'Document Date':'CLE_Booking_Date',
'Currency Code':'Currency_Code',
'Document Type':'Document_Type',
'Transaction No_':'Transaction_No',
'Source Code':'Source_Code',
'User ID':'User_ID',
'Credit Amount':'Credit_Amount',
'Debit Amount':'Debit_Amount',
'Amount (LCY)':'Remaining_Amount',
'Entry Type':'Entry_Type',
'Entry_No':'CLE_No',
'Entry No_':'DCLE_No',
'Customer No_':'Sell_To_Customer_No'
}, inplace = True)
```

```python
Left_Join_DCLE_CLE.info()
```

```python
#add new column to similer values
Left_Join_DCLE_CLE['Link_Customer'] = Left_Join_DCLE_CLE.loc[: , 'Sell_To_Customer_No']
```

```python
Left_Join_DCLE_CLE['CLE_Posting_Date'] = Left_Join_DCLE_CLE.loc[: , 'Posting_Date']
```

```python
Left_Join_DCLE_CLE['Link_Date'] = Left_Join_DCLE_CLE.loc[: , 'Posting_Date']
```

```python
Left_Join_DCLE_CLE
```

```python
#Seprating date from table
#Left_Join_DCLE_CLE['Dates'] = pd.to_datetime(Left_Join_DCLE_CLE['DCLE_Posting_Date']).dt.date
Left_Join_DCLE_CLE["Link_Date"] = pd.to_datetime(Left_Join_DCLE_CLE["Link_Date"])
```

```python
# Replace null values with None
Left_Join_DCLE_CLE['Original_Amount'] = Left_Join_DCLE_CLE['Original_Amount'].where(pd.notnull(Left_Join_DCLE_CLE['Original_Amount']), None)
```
